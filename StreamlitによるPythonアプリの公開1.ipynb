##●"こんにちは"を表示するだけのアプリ
`%%writefile`のマジックコマンドを使って、以下のコードを「app.py」に書き込みます。
<font color ='red'>マジックコマンドは Colaboratory内で利用可能なコマンドです．マジックコマンドを利用するには先頭に %% を入力します．</font>

```python
%%writefile app.py
# 以下を「app.py」に書き込み
import streamlit as st

st.title("こんにちは")
```

## ●"こんにちは"を表示するアプリの起動と動作確認
